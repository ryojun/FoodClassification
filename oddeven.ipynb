```python
def cutUrl(input):
    postion = input.split('/')
    for x in postion:
        if len(x) > 50:
            posis = x
            findName = posis.split('.')
            return findName[0]

def obbEvev(input):
    if int(input)%2 == 0 :
        return 'even'
    elif int(input)%2 != 0:
        return 'odd'
```
